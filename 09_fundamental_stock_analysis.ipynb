```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

#### Extract Fundamentals data for a particular ticker
---

```python
# Getting financial information
yf.Ticker('AAPL')
```

```python
yf.Ticker('AAPL').info
```

```python
tickers = ['AAPL', 'MSFT', 'IBM', 'FB', 'AMZN', 'GOOGL', 'NFLX']
```

```python
infos = []

for x in tickers:
    infos.append(yf.Ticker(x).info)
```

```python
# Define fundamentals we are interested in...
fundamentals = ['dividendYield', 'marketCap', 'beta', 'forwardPE']
```

```python
df = pd.DataFrame(infos)
df.head()
```

```python
# Add company ticker as index
df = df.set_index('symbol')
df
```

```python
# Filter for fundamentals of interest
df.columns[df.columns.isin(fundamentals)]
```

```python
df[df.columns[df.columns.isin(fundamentals)]]
```

```python
# Plot dividend yield
plt.bar(df.index, df.dividendYield, color=('k', 'orange', 'cyan'))
plt.show()
```

```python
plt.bar(df.index, df.forwardPE)
plt.show()
```

```python
# Lets do this for the DJIA
df2 = pd.read_html('https://en.wikipedia.org/wiki/Dow_Jones_Industrial_Average')[1]
df2
```

```python
# Grab the tickers
tickers = df2.Symbol.to_list()
tickers
```

```python
infos = []

for i in tickers:
    infos.append(yf.Ticker(i).info)
```

```python
df2 = pd.DataFrame(infos)
df2 = df2.set_index('symbol')
```

```python
df2
```

```python
df2['dividendYield'].nlargest(5)
```

```python
plt.figure(figsize=(14, 8))
plt.bar(df2['shortName'], df2['dividendYield'])
plt.xticks(rotation=90)
plt.title('DJIA Components Dividend Yield')
plt.show()
```
